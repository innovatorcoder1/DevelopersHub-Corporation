### Task-1

Exploring and visualizing a simple Dataset

```python
# add necessary libraries
from sklearn.datasets import load_iris
import pandas as pd

# load the iris dataset
iris = load_iris()

# create a DataFrame from the iris dataset
dataset = pd.DataFrame(data=iris.data, columns=iris.feature_names)
# add the target variable to the DataFrame
dataset['target'] = iris.target
# display the first few rows of the dataset
print(dataset.head())
```

```python
# print the shape of the dataset
print("Shape of the dataset:", dataset.shape)
```

```python
# print column names of the dataset
print("Column names:", dataset.columns.tolist())
```

```python
# print info about the dataset
print("Dataset info:", dataset.info())
```

```python
# describe the dataset
print("Dataset description:\n", dataset.describe())
```

```python
# create a scatter plot to show relationships between features
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(dataset, hue='target', markers=["o", "s", "D"])
plt.suptitle("Iris Dataset Pairplot", y=1.02)
plt.show()
```

```python
# Use histograms to show value distributions
for feature in iris.feature_names:
    plt.figure(figsize=(8, 4))
    sns.histplot(dataset[feature], kde=True, bins=20)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# use box plots to identify outliers
for feature in iris.feature_names:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='target', y=feature, data=dataset)
    plt.title(f'Box plot of {feature} by target class')
    plt.xlabel('Target Class')
    plt.ylabel(feature)
    plt.show()
```
